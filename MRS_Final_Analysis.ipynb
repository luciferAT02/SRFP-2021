```python
from astropy.io import fits
import matplotlib.pyplot as plt
import numpy as np
import os
import laspec
from laspec import line_index
from scipy.constants import c
from scipy.signal import argrelextrema, peak_widths, find_peaks
from scipy.interpolate import UnivariateSpline
from laspec import normalization
import matplotlib.backends.backend_pdf as pdf

obj = '23_J061149.08+224932.7'
pdff = pdf.PdfPages("Final_MRS_Spectra_"+obj+".pdf")
path2 = '/home/luciferat02/Documents/SRFP_2021/MRS_LAMOST/Objects/'+obj
os.listdir(path2)
print(obj)
for spec_fp in os.listdir(path2):
    with fits.open(path2+'/'+spec_fp) as hdu:
        rband = hdu[2].data
    rflux = rband['FLUX']
    rwv = 10**(rband['LOGLAM'])
    range1 = len(rwv)
    num = 0
    l = 6560
    h = 6570
    r = normalization.normalize_spectrum_general(rwv,rflux, p=1e-6, q=0.1, niter = 100)
    r = np.array([r])
    for i in rwv:
        if i>=l and i<=h:
            num += 1
    fl = np.zeros(num)
    wl = np.zeros(num)
    j = 0
    for i in range(range1):
        if rwv[i]>=l and rwv[i]<=h:
            wl[j] = rwv[i]
            fl[j] = r[0][0][i]
            j += 1
    mean = sum(wl)/len(wl)
    var = sum((wl-mean)**2)/len(wl)
    sigma = np.sqrt(var)
    left = mean - 3*sigma
    right = mean + 3*sigma
    x = np.linspace(l, h, num)
    spline = UnivariateSpline(x, fl-np.max(fl)/2)
    print(spline.roots())
    i = np.argmax(fl)
    wav = wl[i]
    line_dict_info = {
                'line_center': wl[i],
                'line_range': (left,right),
                'line_shoulder_left': (left-0.2,left),
                'line_shoulder_right': (right,right+0.2)}
    EW = line_index.measure_line_index(rwv,r[0][0],flux_err=None, mask=None, 
                                   z=None, line_info = line_dict_info, num_refit=(100, None), 
                                   filepath=None, return_type='dict', verbose=False)
    E_W = EW['EW_int']

    if len(spline.roots())==2:
        print(spec_fp)
        print("Equivalent Width = ",EW['EW_int'])
        fig, ax1 = plt.subplots(1,1, figsize = [15,9])
        ax1.set_xlabel('Wavelength',fontsize = 15)
        ax1.set_ylabel('Normalized Flux', fontsize = 15)
        ax1.set_title('H_alpha Profile of '+obj , fontsize = 15)
        ax1.set_xlim(6550,6580)
        r1, r2 = spline.roots()
        del_lambda = r2-r1
        lambda_halpha = wl[i]
        #denominator = 2*lambda_halpha*np.sqrt(np.log(2))
        #velocity = int(round((c*(del_lambda/denominator))/1000))
        velocity = int(round(((del_lambda/wav)*c)/1000))
        print("Velocity = ",velocity)
        ax1.plot(rwv,r[0][0])
        ax1.plot([left,right],[1,1],linewidth=0.5, c='r',label='EW =%.2f $\AA$'%E_W)
        ax1.plot([r1,r2],[np.max(fl)/2,np.max(fl)/2],linewidth=0.5, c='k',label='FWHM=%.2f $\AA$, Vsin i=%d $kms^{-1}$'%(del_lambda,velocity))
        ax1.legend(loc='best')
        pdff.savefig(fig)
pdff.close()
    
```
