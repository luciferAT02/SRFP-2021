```python
from astropy.io import fits
import matplotlib.pyplot as plt
import numpy as np
import os
import laspec
from laspec import line_index
from scipy.constants import c
from scipy.signal import argrelextrema, peak_widths, find_peaks
from scipy.interpolate import UnivariateSpline
from laspec import normalization
import matplotlib.backends.backend_pdf as pdf
import csv

obj = '24_J062104.61+221010.1'
pdff = pdf.PdfPages("MRS_Spectra_"+obj+".pdf")
path2 = '/home/luciferat02/Documents/SRFP_2021/MRS_LAMOST/Objects/'+obj
os.listdir(path2)
print(obj)
for spec_fp in os.listdir(path2):
    with fits.open(path2+'/'+spec_fp) as hdu:
        rband = hdu[2].data
    rflux = rband['FLUX']
    rwv = 10**(rband['LOGLAM'])
    range1 = len(rwv)
    num = 0
    for i in rwv:
        if i>=6560 and i<=6570:
            num += 1
            
    fl = np.zeros(num)
    wl = np.zeros(num)
    j = 0
    for i in range(range1):
        if rwv[i]>=6560 and rwv[i]<=6570:
            wl[j] = rwv[i]
            fl[j] = rflux[i]
            j += 1
    
    r = normalization.normalize_spectrum_general(rwv,rflux, p=1e-6, q=0.1, niter = 100)
    r = np.array([r])
    x = np.linspace(6560, 6570, num)
    spline = UnivariateSpline(x, fl-np.max(fl)/2)
    i = np.argmax(fl)
    wav = wl[i]
    line_dict_info = {
                'line_center': wl[i],
                'line_range': (wl[i]-2,wl[i]+2),
                'line_shoulder_left': (wl[i]-3,wl[i]-2),
                'line_shoulder_right': (wl[i]+2,wl[i]+3)}
    EW = line_index.measure_line_index(rwv,r[0][0],flux_err=None, mask=None, 
                                   z=None, line_info = line_dict_info, num_refit=(100, None), 
                                   filepath=None, return_type='dict', verbose=False)
    #print(EW['EW_int'])
    print(spline.roots())
    E_W = EW['EW_int']

    if len(spline.roots())==2:
        print("Equivalent Width = ",EW['EW_int'])
        print(spec_fp)
        fig, ax1 = plt.subplots(1,1, figsize = [15,9])
        ax1.set_xlabel('Wavelength',fontsize = 15)
        ax1.set_ylabel('Normalized Flux', fontsize = 15)
        ax1.set_title('H_alpha Profile of '+obj , fontsize = 15)
        ax1.set_xlim(6550,6580)
        r1, r2 = spline.roots()
        del_lambda = r2-r1
        lambda_halpha = wl[i]
        velocity = int(round(((del_lambda/6563)*c)/1000))
        print("Velocity = ",velocity)
       # ax1.text(6563.1,1,"Rotational velocity = (%s) km/s"%velocity)
        ax1.plot(rwv,r[0][0])
        pdff.savefig(fig)
pdff.close()
    
```
